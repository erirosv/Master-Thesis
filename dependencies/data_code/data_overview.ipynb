```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
PATH = 'data/'
path = os.path.abspath(PATH)
csv_files = glob.glob(os.path.join(path, '*.csv'))
print(f'PATH: {path}')
for i in csv_files:
    print(f'CSV_FILES: {i}')
```

## Data Exploration

```python
for file_path in csv_files:
    dataset_name = os.path.splitext(os.path.basename(file_path))[0]
    df = pd.read_csv(file_path)
    label_counts = df['label'].value_counts()
    
    if 1 in label_counts:
        count_1 = label_counts[1]
        print(f"Count of label 1 in {dataset_name}: {count_1}")
    else:
        print(f"Label 1 does not exist in {dataset_name}.")
    
    if 0 in label_counts:
        count_0 = label_counts[0]
        print(f"Count of label 0 in {dataset_name}: {count_0}")
    else:
        print(f"Label 0 does not exist in {dataset_name}.")
    print('----------------------------')
```

## Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for file in csv_files:
    dataset_name = os.path.basename(file)  # Extract the dataset name from the file path
    print(f'Processing dataset: {dataset_name}')
    
    # Load the dataset
    df = pd.read_csv(file)
    
    # Perform any preprocessing or transformations on the dataset if needed
    
    # Create the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(df, cmap='coolwarm', cbar=True)
    
    # Set labels for x-axis and y-axis
    plt.xlabel('Samples')
    plt.ylabel('Genes')
    
    # Set the title of the heatmap
    plt.title(f'Gene Expression Heatmap - {dataset_name}')
    
    # Show or save the heatmap
    plt.show()
    # Alternatively, you can save the heatmap using plt.savefig('heatmap.png') or a specific file path
    
    print(f'Finished processing dataset: {dataset_name}')
```
