```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
data_list_numbers = ['15', '20', '25', '30', '35', '40', '45', '50']
data_list_name = 'stall_limit_'
```

```python
dataframes = []
for number in data_list_numbers:
    file_pattern = f"../data/sensitivity/stall_limit/{data_list_name}{number}/gene_fsr_results_cv_alon.csv"
    print(file_pattern)
    matching_files = glob.glob(file_pattern)
    
    for file in matching_files:
        df = pd.read_csv(file)
        dataframes.append(df)
len(dataframes)
```

```python
df_1 = dataframes[0]
df_2 = dataframes[1]
df_3 = dataframes[2]
df_4 = dataframes[3]
df_5 = dataframes[4]
df_6 = dataframes[5]
df_7 = dataframes[6]
df_8 = dataframes[7]
```

```python
wrapper_names = set(df_1['_wrapper'])
```

```python
# Splitting data in separate variables
df_1_DT = df_1.loc[df_1['_wrapper'] == 'DT', 'cv_SPFSR']
df_1_KNN = df_1.loc[df_1['_wrapper'] == 'KNN', 'cv_SPFSR']
df_1_NB = df_1.loc[df_1['_wrapper'] == 'NB', 'cv_SPFSR']
df_1_SVM = df_1.loc[df_1['_wrapper'] == 'SVM', 'cv_SPFSR']

df_2_DT = df_2.loc[df_2['_wrapper'] == 'DT', 'cv_SPFSR']
df_2_KNN = df_2.loc[df_2['_wrapper'] == 'KNN', 'cv_SPFSR']
df_2_NB = df_2.loc[df_2['_wrapper'] == 'NB', 'cv_SPFSR']
df_2_SVM = df_2.loc[df_2['_wrapper'] == 'SVM', 'cv_SPFSR']

df_3_DT = df_3.loc[df_3['_wrapper'] == 'DT', 'cv_SPFSR']
df_3_KNN = df_3.loc[df_3['_wrapper'] == 'KNN', 'cv_SPFSR']
df_3_NB = df_3.loc[df_3['_wrapper'] == 'NB', 'cv_SPFSR']
df_3_SVM = df_3.loc[df_3['_wrapper'] == 'SVM', 'cv_SPFSR']

df_4_DT = df_4.loc[df_4['_wrapper'] == 'DT', 'cv_SPFSR']
df_4_KNN = df_4.loc[df_4['_wrapper'] == 'KNN', 'cv_SPFSR']
df_4_NB = df_4.loc[df_4['_wrapper'] == 'NB', 'cv_SPFSR']
df_4_SVM = df_4.loc[df_4['_wrapper'] == 'SVM', 'cv_SPFSR']

df_5_DT = df_5.loc[df_5['_wrapper'] == 'DT', 'cv_SPFSR']
df_5_KNN = df_5.loc[df_5['_wrapper'] == 'KNN', 'cv_SPFSR']
df_5_NB = df_5.loc[df_5['_wrapper'] == 'NB', 'cv_SPFSR']
df_5_SVM = df_5.loc[df_5['_wrapper'] == 'SVM', 'cv_SPFSR']

df_6_DT = df_6.loc[df_6['_wrapper'] == 'DT', 'cv_SPFSR']
df_6_KNN = df_6.loc[df_6['_wrapper'] == 'KNN', 'cv_SPFSR']
df_6_NB = df_6.loc[df_6['_wrapper'] == 'NB', 'cv_SPFSR']
df_6_SVM = df_6.loc[df_6['_wrapper'] == 'SVM', 'cv_SPFSR']

df_7_DT = df_7.loc[df_7['_wrapper'] == 'DT', 'cv_SPFSR']
df_7_KNN = df_7.loc[df_7['_wrapper'] == 'KNN', 'cv_SPFSR']
df_7_NB = df_7.loc[df_7['_wrapper'] == 'NB', 'cv_SPFSR']
df_7_SVM = df_7.loc[df_7['_wrapper'] == 'SVM', 'cv_SPFSR']

df_8_DT = df_8.loc[df_8['_wrapper'] == 'DT', 'cv_SPFSR']
df_8_KNN = df_8.loc[df_8['_wrapper'] == 'KNN', 'cv_SPFSR']
df_8_NB = df_8.loc[df_8['_wrapper'] == 'NB', 'cv_SPFSR']
df_8_SVM = df_8.loc[df_8['_wrapper'] == 'SVM', 'cv_SPFSR']
```
