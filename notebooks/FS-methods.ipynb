```python
#Libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.rcParams['font.size'] = 18
```

```python
data =[14, 7, 5, 4]
freq_series = pd.Series(data)
labels = ['Filter', 'Embedded', 'Wrapper', 'Hybrid']
```

```python
plt.figure(figsize=(12, 8))
ax = freq_series.plot(kind="bar")
ax.set_title("FS approach", fontsize=25)
ax.set_xlabel("Individual FS approach", fontsize=22)
ax.set_ylabel("Amount of individual FS approach", fontsize=22)
plt.xticks(rotation=45)
ax.set_xticklabels(labels)
plt.show()
```

```python
total = sum(data)
percentages = [(d / total) * 100 for d in data]

fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(labels, percentages)
#ax.set_title("FS approach", fontsize=25)
ax.set_xlabel("Individual FS approach", fontsize=22)
ax.set_ylabel("Percentage of individual FS approach", fontsize=22)
ax.yaxis.set_major_formatter(ticker.PercentFormatter())
ax.set_ylim(0, 100)
plt.xticks(rotation=45)

label_percentages = {label: percentage for label, percentage in zip(labels, percentages)}
#print(label_percentages)
rounded_dict = {k: round(v, 2) for k, v in label_percentages.items()}
#print(rounded_dict)

for i, (label, percentage) in enumerate(rounded_dict.items()):
    plt.text(i, percentage, str(percentage) + '%', ha='center', va='bottom', fontsize=20)

plt.show()
```
